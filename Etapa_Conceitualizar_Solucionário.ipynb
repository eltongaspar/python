## **Atividade de experimentação 24**

<p align="justify">
Dica: Utilize a "Condição de existência de um triângulo": Para que três segmentos de retas possam formar um triângulo, a soma dos comprimentos de quaisquer dois lados deve ser sempre maior que o comprimento do terceiro lado.
Seja a, b, e c os comprimentos dos três lados. O triângulo existirá se:
<p>


$$a + b > c$$
$$a + c > b$$
$$b + c > a$$

<p align="justify">
Se todas essas condições forem satisfeitas, os lados podem formar um triângulo.
Determinação do Tipo de Triângulo:
<p>
Triângulo Equilátero: Todos os três lados são iguais
$$a = b = c$$
Triângulo Isósceles: Dois lados são iguais e um é diferente $$a = b ≠ c$$ $$a = c ≠ b$$  $$b = c ≠ a$$
Triângulo Escaleno: Todos os lados são diferentes
$$a ≠ b ≠ c$$

```python
l1 = float(input("Digite o comprimento do lado 1: "))
l2 = float(input("Digite o comprimento do lado 2: "))
l3 = float(input("Digite o comprimento do lado 3: "))

# Verifica se os lados formam um triângulo
if l1 + l2 > l3 and l1 + l3 > l2 and l2 + l3 > l1:
    # Determina o tipo do triângulo
    if l1 == l2 == l3:
        print("O triângulo é equilátero.")
    elif l1 == l2 or l1 == l3 or l2 == l3:
        print("O triângulo é isósceles.")
    else:
        print("O triângulo é escaleno.")
else:
    print("Os valores fornecidos não formam um triângulo.")
```

## **Atividade de experimentação 26**

```python
def calcular_fatorial(n):
    """Calcula o fatorial de um número n."""
    if n == 0 or n == 1:
        return 1
    else:
      fat = n * calcular_fatorial(n - 1)
      return fat

numero = int(input("Digite um número para calcular o fatorial: "))

# Calcula e exibe o fatorial do número
fatorial = calcular_fatorial(numero)
print(f"O fatorial de {numero} é {fatorial}.")
```

## **Atividade de experimentação 27**

```python
def soma(a, b):
    return a + b

def subtracao(a, b):
    return a - b

def multiplicacao(a, b):
    return a * b

def divisao(a, b):
    if b != 0:
        return a / b
    else:
        return "Divisão por zero não é permitida"


# programa principal
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))

# Solicita ao usuário escolher a operação
operacao = input("Escolha a operação (Digite + para soma, - para subtração, * para multiplicação, / para divisão): ")

# Executa a operação escolhida
if operacao == '+':
    resultado = soma(num1, num2)
elif operacao == '-':
    resultado = subtracao(num1, num2)
elif operacao == '*':
    resultado = multiplicacao(num1, num2)
elif operacao == '/':
    resultado = divisao(num1, num2)
else:
    resultado = "Operação inválida"

print("Resultado:", resultado)
```

```python
import math

def calcular_hipotenusa(cateto_a, cateto_b):
    return math.sqrt(cateto_a**2 + cateto_b**2)

def calcular_seno(cateto_oposto, hipotenusa):
    return cateto_oposto / hipotenusa

def calcular_cosseno(cateto_adjacente, hipotenusa):
    return cateto_adjacente / hipotenusa

def calcular_tangente(cateto_oposto, cateto_adjacente):
    return cateto_oposto / cateto_adjacente

# Solicitar ao usuário os valores dos catetos
cateto_a = float(input("Digite o comprimento do cateto a: "))
cateto_b = float(input("Digite o comprimento do cateto b: "))

# Calcular e imprimir os resultados
hipotenusa = calcular_hipotenusa(cateto_a, cateto_b)
seno = calcular_seno(cateto_a, hipotenusa)
cosseno = calcular_cosseno(cateto_b, hipotenusa)
tangente = calcular_tangente(cateto_a, cateto_b)

print(f"Hipotenusa: {hipotenusa}")
print(f"Seno: {seno}")
print(f"Cosseno: {cosseno}")
print(f"Tangente: {tangente}")
```

```python
def e_palindromo(texto):
    # Remover espaços e converter para minúsculas para uma comparação uniforme
    texto_limpo = texto.replace(" ", "").lower()

    # Comparar a string com a sua versão invertida
    return texto_limpo == texto_limpo[::-1]

# Testando a função
texto_teste = input('Insira a palavra que iremos verificar: ')
if e_palindromo(texto_teste):
    print(f"'{texto_teste}' é um palíndromo.")
else:
    print(f"'{texto_teste}' não é um palíndromo.")
```

## **Atividade de experimentação 57**

```python
import matplotlib.pyplot as plt
import random

# Gerando dados para os gráficos
# Gráfico de Linha
x_linha = list(range(1, 6))
y_linha = [x**2 for x in x_linha]

# Gráfico de Barras
categorias_barra = ['A', 'B', 'C', 'D', 'E']
valores_barra = [random.randint(1, 10) for _ in categorias_barra]

# Gráfico de Dispersão
x_scatter = [random.random() for _ in range(20)]
y_scatter = [random.random() for _ in range(20)]

# Histograma
dados_histograma = [random.gauss(0, 1) for _ in range(50)]

# Criando a figura e os subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Gráfico de Linha
axs[0, 0].plot(x_linha, y_linha)
axs[0, 0].set_title('Gráfico de Linha')

# Gráfico de Barras
axs[0, 1].bar(categorias_barra, valores_barra, color='skyblue')
axs[0, 1].set_title('Gráfico de Barras')

# Gráfico de Dispersão
axs[1, 0].scatter(x_scatter, y_scatter, color='red')
axs[1, 0].set_title('Gráfico de Dispersão')

# Histograma
axs[1, 1].hist(dados_histograma, bins=10, color='green')
axs[1, 1].set_title('Histograma')

# Ajustando o layout
plt.tight_layout()
plt.show()
```
