## **Atividade de experimentao 24**

<p align="justify">
Dica: Utilize a "Condio de existncia de um tringulo": Para que trs segmentos de retas possam formar um tringulo, a soma dos comprimentos de quaisquer dois lados deve ser sempre maior que o comprimento do terceiro lado.
Seja a, b, e c os comprimentos dos trs lados. O tringulo existir se:
<p>


$$a + b > c$$
$$a + c > b$$
$$b + c > a$$

<p align="justify">
Se todas essas condies forem satisfeitas, os lados podem formar um tringulo.
Determinao do Tipo de Tringulo:
<p>
Tringulo Equiltero: Todos os trs lados so iguais
$$a = b = c$$
Tringulo Issceles: Dois lados so iguais e um  diferente $$a = b  c$$ $$a = c  b$$  $$b = c  a$$
Tringulo Escaleno: Todos os lados so diferentes
$$a  b  c$$

```python
l1 = float(input("Digite o comprimento do lado 1: "))
l2 = float(input("Digite o comprimento do lado 2: "))
l3 = float(input("Digite o comprimento do lado 3: "))

# Verifica se os lados formam um tringulo
if l1 + l2 > l3 and l1 + l3 > l2 and l2 + l3 > l1:
    # Determina o tipo do tringulo
    if l1 == l2 == l3:
        print("O tringulo  equiltero.")
    elif l1 == l2 or l1 == l3 or l2 == l3:
        print("O tringulo  issceles.")
    else:
        print("O tringulo  escaleno.")
else:
    print("Os valores fornecidos no formam um tringulo.")
```

## **Atividade de experimentao 26**

```python
def calcular_fatorial(n):
    """Calcula o fatorial de um nmero n."""
    if n == 0 or n == 1:
        return 1
    else:
      fat = n * calcular_fatorial(n - 1)
      return fat

numero = int(input("Digite um nmero para calcular o fatorial: "))

# Calcula e exibe o fatorial do nmero
fatorial = calcular_fatorial(numero)
print(f"O fatorial de {numero}  {fatorial}.")
```

## **Atividade de experimentao 27**

```python
def soma(a, b):
    return a + b

def subtracao(a, b):
    return a - b

def multiplicacao(a, b):
    return a * b

def divisao(a, b):
    if b != 0:
        return a / b
    else:
        return "Diviso por zero no  permitida"


# programa principal
num1 = float(input("Digite o primeiro nmero: "))
num2 = float(input("Digite o segundo nmero: "))

# Solicita ao usurio escolher a operao
operacao = input("Escolha a operao (Digite + para soma, - para subtrao, * para multiplicao, / para diviso): ")

# Executa a operao escolhida
if operacao == '+':
    resultado = soma(num1, num2)
elif operacao == '-':
    resultado = subtracao(num1, num2)
elif operacao == '*':
    resultado = multiplicacao(num1, num2)
elif operacao == '/':
    resultado = divisao(num1, num2)
else:
    resultado = "Operao invlida"

print("Resultado:", resultado)
```

```python
import math

def calcular_hipotenusa(cateto_a, cateto_b):
    return math.sqrt(cateto_a**2 + cateto_b**2)

def calcular_seno(cateto_oposto, hipotenusa):
    return cateto_oposto / hipotenusa

def calcular_cosseno(cateto_adjacente, hipotenusa):
    return cateto_adjacente / hipotenusa

def calcular_tangente(cateto_oposto, cateto_adjacente):
    return cateto_oposto / cateto_adjacente

# Solicitar ao usurio os valores dos catetos
cateto_a = float(input("Digite o comprimento do cateto a: "))
cateto_b = float(input("Digite o comprimento do cateto b: "))

# Calcular e imprimir os resultados
hipotenusa = calcular_hipotenusa(cateto_a, cateto_b)
seno = calcular_seno(cateto_a, hipotenusa)
cosseno = calcular_cosseno(cateto_b, hipotenusa)
tangente = calcular_tangente(cateto_a, cateto_b)

print(f"Hipotenusa: {hipotenusa}")
print(f"Seno: {seno}")
print(f"Cosseno: {cosseno}")
print(f"Tangente: {tangente}")
```

```python
def e_palindromo(texto):
    # Remover espaos e converter para minsculas para uma comparao uniforme
    texto_limpo = texto.replace(" ", "").lower()

    # Comparar a string com a sua verso invertida
    return texto_limpo == texto_limpo[::-1]

# Testando a funo
texto_teste = input('Insira a palavra que iremos verificar: ')
if e_palindromo(texto_teste):
    print(f"'{texto_teste}'  um palndromo.")
else:
    print(f"'{texto_teste}' no  um palndromo.")
```

```python
# Nome do arquivo
nome_arquivo = 'meu_arquivo.txt'

try:
    with open(nome_arquivo, 'r') as arquivo:
        linhas = arquivo.readlines()

    # Inicializando contadores
    numero_linhas = len(linhas)
    numero_palavras = 0
    numero_caracteres = 0

    # Contando palavras e caracteres
    for linha in linhas:
        palavras = linha.split()
        numero_palavras += len(palavras)
        numero_caracteres += len(linha)

    # Exibindo as contagens
    print(f"Contagem no arquivo {nome_arquivo}:")
    print(f"Linhas: {numero_linhas}")
    print(f"Palavras: {numero_palavras}")
    print(f"Caracteres: {numero_caracteres}")
except FileNotFoundError:
    print(f"O arquivo {nome_arquivo} no foi encontrado.")
```

## **Atividade de experimentao 57**

```python
import matplotlib.pyplot as plt
import random

# Gerando dados para os grficos
# Grfico de Linha
x_linha = list(range(1, 6))
y_linha = [x**2 for x in x_linha]

# Grfico de Barras
categorias_barra = ['A', 'B', 'C', 'D', 'E']
valores_barra = [random.randint(1, 10) for _ in categorias_barra]

# Grfico de Disperso
x_scatter = [random.random() for _ in range(20)]
y_scatter = [random.random() for _ in range(20)]

# Histograma
dados_histograma = [random.gauss(0, 1) for _ in range(50)]

# Criando a figura e os subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Grfico de Linha
axs[0, 0].plot(x_linha, y_linha)
axs[0, 0].set_title('Grfico de Linha')

# Grfico de Barras
axs[0, 1].bar(categorias_barra, valores_barra, color='skyblue')
axs[0, 1].set_title('Grfico de Barras')

# Grfico de Disperso
axs[1, 0].scatter(x_scatter, y_scatter, color='red')
axs[1, 0].set_title('Grfico de Disperso')

# Histograma
axs[1, 1].hist(dados_histograma, bins=10, color='green')
axs[1, 1].set_title('Histograma')

# Ajustando o layout
plt.tight_layout()
plt.show()
```
